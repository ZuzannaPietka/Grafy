```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
print('NetworkX version: {}'.format(nx.__version__))
```

```python
G = nx.florentine_families_graph()
print(nx.info(G))
```

```python
pos_spring = nx.spring_layout(G)
nx.draw(G, pos_spring, with_labels=True)
```

```python
from operator import itemgetter
degc = nx.degree_centrality(G)
# let's list the scores
sorted(degc.items(), key=itemgetter(1), reverse=True)
```

```python
bet = nx.betweenness_centrality(G)
sorted(bet.items(), key=itemgetter(1), reverse=True)
```

```python
clos = nx.closeness_centrality(G)
sorted(clos.items(), key=itemgetter(1), reverse=True)
```

Najpierw tworzymy tablicę z wartościami dla poszczególnych węzłów.
Uwaga: `degc` jest słownikiem.

```python
nodes = G.nodes()
values = [degc.get(node) for node in nodes]
values
```

Następnie należy te wartości "przełożyć" na kolory za pomocą colormap.

```python
ec = nx.draw_networkx_edges(G, pos_spring, alpha=0.2)
nc = nx.draw_networkx_nodes(G, pos_spring, nodelist=nodes, node_color=values, 
                            with_labels=True, cmap=plt.cm.magma_r)
plt.colorbar(nc)
plt.axis('off')
plt.show()
```

```python
values = [bet.get(node) for node in nodes]
ec = nx.draw_networkx_edges(G, pos_spring, alpha=0.2)
nc = nx.draw_networkx_nodes(G, pos_spring, nodelist=nodes, node_color=values, 
                            with_labels=True, cmap=plt.cm.viridis_r)
plt.colorbar(nc)
plt.axis('off')
plt.show()
```
